```python
fulldf=pd.read_csv('cleanedFulldf.csv', index_col=0)
fulldf1=fulldf.copy()
fulldf1['dateutc'] = fulldf1['dateutc'].astype('datetime64')
fulldf1['dateloaded'] = fulldf1['dateloaded'].astype('datetime64')
newdata=fulldf1[fulldf1['dateloaded']=='2015-12-06'].copy()
olddata=fulldf1[fulldf1['dateloaded']=='2015-11-29'].copy()
olddata=olddata[olddata['dateutc']>='2015-12-06'].copy()
lol=newdata[['lowest_price_good','id','dateutc']].copy()
newdf=pd.merge(lol, olddata, how='inner', on=['dateutc','id'],suffixes=('_now', '_ago'))
```
